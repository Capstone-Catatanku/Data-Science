```python
import pandas as pd

# ============================================
# DATA DICTIONARY - KEUANGAN PRIBADI
# ============================================

# Sheet 1: Informasi Umum
df_info = pd.DataFrame([
    ["Nama Dataset", "Data-clean.csv"],
    ["Jumlah Kolom", 5],
    ["Jumlah Baris", 4548],
    ["Periode Data", "2019-12-28 s/d 2024-12-31"],
    ["Sumber Data", "Catatan keuangan pribadi"],
    ["Tanggal Dibuat", "2024-12-31"]
])

# Sheet 2: Daftar Kolom
df_kolom = pd.DataFrame([
    [1, "tanggal", "Date", "Tanggal transaksi (YYYY-MM-DD)"],
    [2, "deskripsi_transaksi", "String", "Nama barang/jasa yang dibeli"],
    [3, "kategori", "Kategorikal", "Pengelompokan jenis transaksi"],
    [4, "nominal", "Float", "Nilai transaksi dalam Rupiah"],
    [5, "tipe_transaksi", "Kategorikal", "Pengeluaran atau Pemasukan"]
], columns=["No", "Nama Kolom", "Tipe Data", "Deskripsi"])

# Sheet 3: Daftar Kategori
df_kategori = pd.DataFrame([
    [1, "Belanja", "Pembelian barang non-makanan", "lego, koper, tas, sepatu, elektronik"],
    [2, "Makanan", "Pembelian makanan/minuman", "gofood, grabfood, jajan, restoran"],
    [3, "Transportasi", "Biaya perjalanan & kendaraan", "bensin, tol, tiket krl, ojol"],
    [4, "Tagihan", "Biaya rutin bulanan", "listrik, air, wifi, BPJS, cicilan"],
    [5, "Tempat Tinggal", "Biaya hunian", "uang kos, deposit, perbaikan rumah"],
    [6, "Kesehatan", "Biaya medis & kebugaran", "obat, cek lab, vitamin, dokter"],
    [7, "Hiburan", "Aktivitas rekreasi", "bioskop, karaoke, bowling, game"],
    [8, "Lain-lain", "Transaksi lain", "parkir, denda, sumbangan, admin bank"],
    [9, "Investasi", "Pembelian aset", "crypto, saham, emas, reksadana"],
    [10, "Pemasukan", "Uang masuk", "refund, cashback, bonus, thr"]
], columns=["No", "Kategori", "Deskripsi", "Contoh"])

# Sheet 4: Statistik Nominal
df_statistik = pd.DataFrame([
    ["Minimum", 100],
    ["Q1 (25%)", 12000],
    ["Median (50%)", 65000],
    ["Q3 (75%)", 200000],
    ["Maksimum", 1025000],
    ["Rata-rata", 156000]
], columns=["Statistik", "Nilai (Rp)"])

# Sheet 5: Tipe Transaksi
df_tipe = pd.DataFrame([
    ["Pengeluaran", "Uang keluar", "MENGURANGI saldo", "99.4% (4520 transaksi)"],
    ["Pemasukan", "Uang masuk", "MENAMBAH saldo", "0.6% (28 transaksi)"]
], columns=["Tipe", "Arti", "Dampak ke Saldo", "Distribusi"])

# Sheet 6: Contoh Data
df_contoh = pd.DataFrame([
    ["2024-12-31", "order gas torch manual", "Belanja", 18900, "Pengeluaran"],
    ["2024-12-30", "beli leivy body wash", "Belanja", 64000, "Pengeluaran"],
    ["2024-12-29", "uang kembalian", "Pemasukan", 713000, "Pemasukan"],
    ["2024-12-28", "beli sepatu new era", "Belanja", 140000, "Pengeluaran"]
], columns=["tanggal", "deskripsi_transaksi", "kategori", "nominal", "tipe_transaksi"])

# ============================================
# SIMPAN KE EXCEL
# ============================================

filename = "Data_Dictionary_Keuangan.xlsx"

with pd.ExcelWriter(filename, engine='openpyxl') as writer:
    df_info.to_excel(writer, sheet_name="Informasi_Umum", index=False, header=False)
    df_kolom.to_excel(writer, sheet_name="Daftar_Kolom", index=False)
    df_kategori.to_excel(writer, sheet_name="Daftar_Kategori", index=False)
    df_statistik.to_excel(writer, sheet_name="Statistik_Nominal", index=False)
    df_tipe.to_excel(writer, sheet_name="Tipe_Transaksi", index=False)
    df_contoh.to_excel(writer, sheet_name="Contoh_Data", index=False)

print(f"✅ BERHASIL! File '{filename}' telah dibuat.")
print("📁 Cek folder tempat file ini dijalankan.")
```
